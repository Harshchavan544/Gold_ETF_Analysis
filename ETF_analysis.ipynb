# **Does ETFs shows the same growth as prices of gold**

```python
from scipy.stats import f_oneway
```

```python
merged=gold.merge(icici,left_on="Date",right_on="Date ",how="inner")   #merging data on date 
```

```python
merged.head() #hence ETF shows the same growth therefore taking icici_etf for analysis
```

```python
merged.rename(columns={"percent_x":"rate_gold"},inplace="True")
merged.rename(columns={"percent_y":"rate_icici"},inplace="True")
```

```python
merged.head()
```

```python
# Ho = Prices of gold are similar as prices of ETF
# Ha = There is noticable change in prices of ETF and gold
```

```python
alpha=0.05
```

```python
p_value = f_oneway(merged["rate_gold"],merged["rate_icici"])[1]
p_value
```

```python
if p_value < alpha:
    print("Reject Ho")
else:
    print("Fail to Reject Ho")
```

```python
# Here p_value is greater then 0.05
# Since the p_vaue is greater then 0.05 therefore we assume there is no noticeable difference in the rates of gold and ETFs
```
